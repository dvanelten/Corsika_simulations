```python
a = list()
b = list()
```

```python
for i in range(len(indices)):            #fuer alle Photonen mit Index i wird nun aus der zugehoerigen Reihe aus result
    print(indices[i])                    #gewichtet mit der Wahrscheinlichkeit der Eintraege ein sipm (dessen response
    #result.iloc[indices[i],:]           #wert)gezogen
    resp = (np.random.choice(result.iloc[indices[i]],p=result.iloc[indices[i],:])) 
    a.append(indices[i])                 #fuege indices der photonen in eine liste
    b.append(resp)                       #fuege random response in zweite liste
    print(resp)
```

```python
print(a)
print(b)
```

```python
df = pd.DataFrame(list(zip(a,b)))
df.rename(columns={0:'photon_index'}, inplace=True)
df.rename(columns={1:'pde'}, inplace=True)
df
```
